```python
import PyPDF2
PDFFile = open('2104.07808.pdf','rb')

link_to_paper = []
PDF = PyPDF2.PdfFileReader(PDFFile)
pages = PDF.getNumPages()
key = '/Annots'
uri = '/URI'
ank = '/A'

for page in range(pages):
    print("Current Page: {}".format(page))
    pageSliced = PDF.getPage(page)
    pageObject = pageSliced.getObject()
    if key in pageObject.keys():
#         break
        ann = pageObject[key]
        for a in ann:
            u = a.getObject()
            print(u)
            if ank not in u.keys():
                continue
            if uri in u[ank].keys():
                print(u[ank][uri])
                link_to_paper.append(u[ank][uri])
```
